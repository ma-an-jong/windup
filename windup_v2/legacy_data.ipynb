```python
import pandas as pd
```

```python
stat = pd.read_csv('data/pitcher_stat.csv')
```

```python
pitcher_list = stat.loc[:,'name':'team'].drop_duplicates(subset=None, keep='first', inplace=False, ignore_index=False).values
```

```python
columns = list(stat.columns)
```

```python
stat.head()
```

```python
average = pd.DataFrame(columns = ['name','team'] + columns[3:])
```

```python
for name,team in pitcher_list:
    tmp = stat.loc[(stat['name']==name) & (stat['team'] == team)]
    row = [name,team]
    for col in columns[3:]:
        for value in tmp[col]:
            total = 0
            try:
                total += float(value)
            except:
                continue
        row.append(total/len(tmp[col]))
    average.loc[len(average)] = row
```

```python
average.to_csv('data/average.csv',index=False)
```
